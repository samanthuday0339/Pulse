## Step 2: Daily Batching

Each calendar day is treated as one batch of reviews, simulating
real-world daily ingestion.


```python
import pandas as pd

df = pd.read_csv("../data/swiggy_reviews_raw.csv")
df["review_date"] = pd.to_datetime(df["review_date"])

daily_batches = {
    date: group["review_text"].tolist()
    for date, group in df.groupby(df["review_date"].dt.date)
}

print("Total daily batches:", len(daily_batches))

```
